```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


test_img_path="..//Cat03.jpg"

# Read the image in grayscale mode
image = cv2.imread(test_img_path, cv2.IMREAD_GRAYSCALE)
image=cv2.resize(image,(500,500))

# Perform histogram equalization
equalized_image = cv2.equalizeHist(image)

# Display the original and equalized images
cv2.imshow('Original Image', image)
cv2.imshow('Histogram Equalized Image', equalized_image)

# Wait for a key press and then close all open windows
cv2.waitKey(0)
cv2.destroyAllWindows()

# Optionally, display the histograms
# Calculate histograms using OpenCV
hist_original = cv2.calcHist([image], [0], None, [256], [0, 256])
hist_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])

plt.figure(figsize=(10, 4))

# Plot original histogram
plt.subplot(1, 2, 1)
plt.plot(hist_original, color='gray')
plt.title('Original Histogram')
plt.xlabel('Bins')
plt.ylabel('Normalized Pixels')

# Plot equalized histogram
plt.subplot(1, 2, 2)
plt.plot(hist_equalized, color='black')
plt.title('Equalized Histogram')
plt.xlabel('Bins')
plt.ylabel('Normalized Pixels')

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
