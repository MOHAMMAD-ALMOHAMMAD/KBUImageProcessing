```python
import cv2 
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

```

```python
#reading the image and showing it
img=cv2.imread("Images/Cat03.jpg")
img=cv2.resize(img,(500,500))
cv2.imshow("test",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#image dimensions
img.shape
```

```python
#creating a matrix density picture

img[100:105,:]=0.5
img[:,100:105]=1
cv2.imshow("test",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#converting rgb image to grayscale

grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("grayscaletest", grayscale)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#cropping an image

cv2.imshow("croptest",img[:250,:250])
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Rotating an image

rotimg=cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
#rotimg=cv2.rotate(rotimg,cv2.ROTATE_90_CLOCKWISE) #rotate by 180 degress
#rotimg=cv2.rotate(rotimg,cv2.ROTATE_90_CLOCKWISE) #rotate by 270 degrees
#rotimg=cv2.rotate(rotimg,cv2.ROTATE_90_CLOCKWISE) #rotate by 360 degrees


cv2.imshow("rotationtest",rotimg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Creating an RGB image histogram to show frequencies of pixel intensity distribution
for i,colour in enumerate(['b','g','r']):
 hist=cv2.calcHist([img],[i],None,[256],[0,256])
 plt.plot(hist,color=colour)

plt.show()
```

```python
#creating a mask

maskimg=np.zeros([500,500,3],dtype=np.uint8)

maskimg[100:200,100:500]=255 #make pixels from (100,100) to (200,500) all white

cv2.imshow("maskimgtest",maskimg*img) #multiply the mask with the original image
cv2.waitKey(0)
cv2.destroyAllWindows()

```
